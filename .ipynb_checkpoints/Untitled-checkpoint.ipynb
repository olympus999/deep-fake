```python
def make_mapper_weather(feature_columns_categorical, feature_columns_continuous):
    
    def adjust_r(r):
        r = r + r_span
        return  r

    mapper = {}
    r = 0
    r_span = 1
    
    for col in feature_columns_categorical:
        mapper[col] = {}
        u = df_train_weather[col].unique()
        for val in u:
            r = adjust_r(r)
            mapper[col][val] = {}
            mapper[col][val]['row'] = r
        cnt = 0
        r = adjust_r(r)
    
    for col in feature_columns_continuous:
        mapper[col] = {}
        mapper[col]['row'] = r
        r = adjust_r(r)
        mapper[col]['None'] = r
        r = adjust_r(r)
    
    return mapper
```

```python
mapper_weather = make_mapper_weather(f_columns_weather_cat, f_columns_weather_continuous)
mapper_weather
```

```python
def data_for_model_sequential_arr(arr, column_mapper, feature_columns_categorical, feature_columns_continuous, mapper, width):
    
    img = np.zeros(width)
    arr_idx = []
    arr_val = []
    for col in feature_columns_categorical:
        r = mapper[col][arr[column_mapper[col]]]['row']
        arr_idx.append(r)
        arr_val.append(1)

    for col in feature_columns_continuous:
        if arr[column_mapper[col]] != 'None':
            r = mapper[col]['row']
            arr_idx.append(r)
            arr_val.append(arr[column_mapper[col]])
        else:
            r = mapper[col]['None']
            arr_idx.append(r)
            arr_val.append(1)
    
    img[arr_idx] = arr_val
    y = arr[column_mapper['air_temperature']]
    return [img, y]
```

```python
target = np.log1p(df_train_joined["meter_reading"])
features = df_train_joined.drop('meter_reading', axis = 1)
```

```python
import lightgbm as lgb
import gc
```

```python
categorical_features = ["building_id", "site_id", "meter", "primary_use", "weekend"]
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
models = []
params = {
    "objective": "regression",
    "boosting": "gbdt",
    "num_leaves": 1280,
    "learning_rate": 0.05,
    "feature_fraction": 0.85,
    "reg_lambda": 2,
    "metric": "rmse",
}
for train_index,test_index in kf.split(features):
    train_features = features.loc[train_index]
    train_target = target.loc[train_index]
    
    test_features = features.loc[test_index]
    test_target = target.loc[test_index]
    
    d_training = lgb.Dataset(train_features, label=train_target,categorical_feature=categorical_features, free_raw_data=False)
    d_test = lgb.Dataset(test_features, label=test_target,categorical_feature=categorical_features, free_raw_data=False)
    
    model = lgb.train(params, train_set=d_training, num_boost_round=1000, valid_sets=[d_training,d_test], verbose_eval=25, early_stopping_rounds=50)
    models.append(model)
    del train_features, train_target, test_features, test_target, d_training, d_test
    gc.collect()
```
